```python
## import csvs of calculated oh reactivities
csv_dir = '../Data for project/'
oh_reac_list = list()
for file in os.listdir(csv_dir):
    oh_reac_list.append(pd.read_csv(csv_dir+file))
```

```python
df_all = pd.concat(oh_reac_list)
df_all
```
